```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import os
import random
```

```python
cmap = plt.get_cmap()
my_cmap = cmap(np.arange(cmap.N))
my_cmap[:,-1] = np.linspace(0, 1, cmap.N)
my_cmap = ListedColormap(my_cmap)
```

```python
path = '../houses/valannot/valannot'
images = os.listdir(path)
```

```python
pic, = random.sample(images, k=1)

img_path = f'../houses/val/val/{pic}'
segment_path = f'../houses/valannot/valannot/{pic}'
```

```python
fig, ax = plt.subplots(figsize = (20,20))

img1 = plt.imread(img_path)
img2 = plt.imread(segment_path)

img = ax.imshow(img1, interpolation='nearest', origin='lower')
img = ax.imshow(img2, 
                interpolation='nearest', 
                alpha=0.5, 
                origin='lower',
                cmap=my_cmap,
               )

ax.set_title(img_path)
fig.show()
```
